```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/datasets/norahalghmmas/dataproject7/day35_project.csv")

plt.figure()
df.iloc[:, 0].hist()
plt.show()

plt.figure()
sns.boxplot(y=df.iloc[:, 0])
plt.show()

plt.figure()
df.iloc[:, 1].hist()
plt.show()

plt.figure()
sns.scatterplot(x=df.iloc[:, 0], y=df.iloc[:, 1])
plt.show()

correlation = df.corr(numeric_only=True)

plt.figure()
sns.heatmap(correlation, annot=True)
plt.show()
```
